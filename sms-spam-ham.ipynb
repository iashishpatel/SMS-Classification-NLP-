### I am going to use this dataset to build a prediction model that will accurately classify which texts are spam

```python
#importing needed Libraries
import pandas as pd
import nltk
```

```python
df = pd.read_csv("../input/sms-spam-collection-dataset/spam.csv",encoding="ISO-8859-1")
df.head()
```

```python
df.drop(["Unnamed: 2","Unnamed: 3","Unnamed: 4"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.rename(columns = {"v1":"label","v2":"sms"},inplace=True)
df.head()
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
wordlem = WordNetLemmatizer()
sms = []
for i in range(len(df["sms"])):
    review = re.sub("[^a-zA-Z]"," ",df["sms"][i])
    review = review.lower()
    review = review.split()
    review = [wordlem.lemmatize(word) for word in review if word not in set(stopwords.words("english"))]
    review = " ".join(review)
    sms.append(review)
```

```python
df.drop(["sms"],axis=1,inplace=True)
df["sms"] = sms
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000)
x = cv.fit_transform(df["sms"]).toarray()
x.shape
```

```python
y=df["label"]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(x_train,y_train)
y_pred = nb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

print(confusion_matrix(y_pred,y_test))
```

```python
acc = accuracy_score(y_pred,y_test)
acc
```
